```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
#read txt sounding files
#path where sounding files are located
folder_dir='/glade/scratch/sisichen/running_jobs/WRF_SNOWIE/Analysis/Sounding_files/'
os.chdir(folder_dir)
Sounding_file_list1=glob.glob('kcrh*')
Sounding_file_list2=glob.glob('KLOW*')
Sounding_file_list3=glob.glob('UIUC_Mobile*.txt')
#read kcrh file
for Sfile in Sounding_file_list1:
    print(Sfile)
    df = pd.read_csv(Sfile, sep='\s+', encoding='latin1',skiprows=2, header=None, names=\
                    ['time_launch','time_utc','geo_height','lat','lon','pres','temp',\
                     'rh','wspd','wdir','slant_rng','ele_ang'])
    df=df.replace({'//': np.NaN})
    df[['geo_height','pres','temp','rh','wspd','wdir','slant_rng','ele_ang']] = \
        df[['geo_height','pres','temp','rh','wspd','wdir','slant_rng','ele_ang']].astype(float)
    #df['td']=df.temp-(100.0-df.rh)/5.#dew point temperature calculation
    pd.options.display.float_format = '{:,.1f}'.format
    
    #df=df.replace({np.nan: -999})
    df.to_csv('formatted_'+Sfile, na_rep='NaN',header=False, index=False, columns= ['geo_height','pres', 'temp', 'rh','wspd','wdir'], sep='\t', mode='w',float_format='%.1f')
#read KLOW file
for Sfile in Sounding_file_list2:
    print(Sfile)
    df = pd.read_csv(Sfile, sep='\s+', encoding='latin1',skiprows=2, header=None, names=\
                    ['time_launch','time_utc','geo_height','lat','lon','pres','temp',\
                     'rh','wspd','wdir','slant_rng','ele_ang'])
    df=df.replace({'//': np.NaN})
    df[['geo_height','pres','temp','rh','wspd','wdir','slant_rng','ele_ang']] = \
        df[['geo_height','pres','temp','rh','wspd','wdir','slant_rng','ele_ang']].astype(float)
    #df['td']=df.temp-(100.0-df.rh)/5.#dew point temperature calculation
    pd.options.display.float_format = '{:,.1f}'.format
    
    #df=df.replace({np.nan: -999})
    df.to_csv('formatted_'+Sfile, na_rep='NaN',header=False, index=False, columns= ['geo_height','pres', 'temp', 'rh','wspd','wdir'], sep='\t', mode='w',float_format='%.1f')
#read UIUC file
for Sfile in Sounding_file_list3:
    print(Sfile)
    df = pd.read_csv(Sfile, sep='\s+', encoding='latin1',skiprows=15, header=None, names=\
                    ['time','pres','temp','td','rh','uvel','vvel','wspd','wdir','wvel','lon','lat',\
                     'ele_ang','hor_ang','geo_height','qp','qt','qrh','qu','qv','qdz'])
    df[['time','pres','temp','td','rh','uvel','vvel','wspd','wdir','wvel','lon','lat',\
                     'ele_ang','hor_ang','geo_height','qp','qt','qrh','qu','qv','qdz']] = \
        df[['time','pres','temp','td','rh','uvel','vvel','wspd','wdir','wvel','lon','lat',\
                     'ele_ang','hor_ang','geo_height','qp','qt','qrh','qu','qv','qdz']].astype(float)
    pd.options.display.float_format = '{:,.1f}'.format
    
    #df=df.replace({np.nan: -999})
    df.to_csv('formatted_'+Sfile, header=False, index=False, columns= ['geo_height','pres', 'temp', 'rh','wspd','wdir'], sep='\t', mode='w',float_format='%.1f')

print('finished')
```
