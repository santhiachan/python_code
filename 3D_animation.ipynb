```python
from netCDF4 import Dataset
from mayavi import mlab
import numpy as np
mlab.init_notebook('png', 400, 300)
#database & file io
import pandas as pd
from io import StringIO
from netCDF4 import Dataset
#import ffmpeg
```

```python
#import dynamic field data
dynancfile='flow_new.nc'
fh = Dataset(dynancfile,mode='r')
ZX=fh.variables['ZX'][:]
ZY=fh.variables['ZY'][:]
ZZ=fh.variables['ZZ'][:]
TH=fh.variables['TH'][:]
QV=fh.variables['QV'][:]
fh.close()
```

```python
#set up mesh grid
x1=np.linspace(0,8.3,64)
y1=np.linspace(0,8.3,64)
z1=np.linspace(0,8.3,64)
yv,zv,xv = np.meshgrid(y1,z1,x1,indexing='ij')
#zx=ZX[1][:][:][:]
#zy=ZY[1][:][:][:]
#zz=ZZ[1][:][:][:]
vort=ZX**2+ZY**2+ZZ**2
```

```python
#import droplet data using pandas
df = pd.read_csv('droplet_new.track',sep='\s+',header=None,names=['time','id','x','y','z','r','s'])
#save it to csv file
#df.to_csv(r'droplet_track.csv')
```

```python
#select data
%matplotlib inline
index_range = range(0,6400) #select droplets
nframes     = range(0,10) #select time frames
field_data=vort #select field data
```

```python
#first set up the figure, axis, & plot element
for i in nframes:
    xdrop=df[df['time']==df.time[i]].x*1e2
    ydrop=df[df['time']==df.time[i]].y*1e2
    zdrop=df[df['time']==df.time[i]].z*1e2
    mlab.clf(figure=None)
    mlab.figure(3,fgcolor=(.0,.0,.0), bgcolor=(1.0,1.0,1.0))
    fig1=mlab.points3d(xdrop,ydrop,zdrop,scale_factor=0.05) #droplets
    fig1=mlab.contour3d(yv,zv,xv,field_data[i][:][:][:],colormap='viridis',contours=10,opacity=0.2)
    mlab.axes(fig1,x_axis_visibility=True,xlabel=' ', ylabel=' ',zlabel=' ')
    mlab.axes.font_factor=0.1
    fig1
    num=str(i).zfill(3)
    figfile="plot"+num+".png"
    mlab.savefig(figfile)
    mlab.clf(figure=None)
    
```
